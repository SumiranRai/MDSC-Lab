```python
import matplotlib.pyplot as plt # from matploblib import pyplot as plt
import numpy as np
```

```python
np.random.seed(208)

mu = 3
n = 50

x = np.random.normal(mu, 1, size =n)
y = np.random.normal(mu, 1, size =n)
z = np.random.normal(mu, 1, size =n)
```

rcParams --> Runtime Configuration Parameters

```python
plt.rcParams["figure.figsize"] = (8, 6)
```

### 3D Scatter Plot

```python
ax = plt.axes(projection = "3d")
ax.scatter3D(x, y, z, s=100);

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.view_init(30, 30);# (Angle of Elevation, Angle of Animuth)
```

```python
type(ax)
```

3D Line Plot

```python
omega = 2
z_line = np.linspace(0, 10, 100)
x_line = np.cos(omega * z_line)
y_line = np.sin(omega * z_line)
```

```python
ax = plt.axes(projection = "3d");
ax.plot(x_line, y_line, z_line, lw = 4)
type(ax)
```

### 3D Surface Plot

Example surface: z = 50 - (x<sup>2</sup> + y<sup>2</sup>)

```python
def function_z(x, y):
    return 50 - (x**2 + y**2)
```

```python
N = 50

x_values = np.linspace(-5, 5, N)
y_values = np.linspace(-5, 5, N)

X, Y = np.meshgrid(x_values, y_values)
```

```python
X.shape
plt.scatter(X, Y)
```

```python
Z = function_z(X, Y)
Z.shape
```

```python
ax = plt.axes(projection = "3d")
ax.plot_surface(X, Y, Z);
```

### 3D Wireframe

```python
N = 10

x_values = np.linspace(-5, 5, N)
y_values = np.linspace(-5, 5, N)

X, Y = np.meshgrid(x_values, y_values)

Z = function_z(X, Y)
ax = plt.axes(projection = "3d")
ax.plot_wireframe(X, Y, Z, color = 'green', lw = 5)
```
