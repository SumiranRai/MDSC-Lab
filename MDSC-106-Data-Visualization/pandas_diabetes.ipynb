# Pandas Assignment on Diabetes Dataset

Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes

# Load the dataset
diabetes = load_diabetes()

# Convert to pandas DataFrame
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
```

```python
df.head()
```

```python
df.info()
df.shape
```

Add 2 synthetic columns with column name 'Name of Patient', 'Patient Aadhar No.'.

```python
df['Name of Patient'] = np.nan
df['Patient Aadhar No.'] = np.nan
df.info()
```

Write meaningful data to those columns ( based on real life examples ).

```python
import random 

first_names = [
    "Sumiran",
    "Ritick Raj",
    "Narendra",
    "Guru",
    "Muthangi",
    "Sairamaraju",
    "Tushar",
    "Shivansh",
    "Nandavardhan"
]

last_names = [
    "Rai",
    "Gupta",
    "Sai",
    "Ranga Sai",
    "Prasanth",
    "Vyasaraju",
    "Sharma",
    "Reddy"
]

df['Name of Patient'] = [f"{random.choice(first_names)} {random.choice(last_names)}" for _ in range(len(df))]

def generate_aadhar():
    return ''.join([str(random.randint(0, 9)) for _ in range(16)])

df['Patient Aadhar No.'] = [generate_aadhar() for _ in range(len(df))]
df.head(10)
```

Check for null values and display the count of null values in the dataframe under each column name.

```python
null_counts = df.isnull().sum()

print("Count of null values in each column:")
print(null_counts)
```

Drop one of the column you have created i.e. 'Name of Patient'.

```python
df.drop(columns=['Name of Patient'], inplace=True)
df.info()
```

Fill the null values in column 1 i.e. 'Age' with meaningful data based on the other values available.

```python
mean_age = df['age'].mean()
df['age'].fillna(mean_age, inplace=True)

df.head()
```
