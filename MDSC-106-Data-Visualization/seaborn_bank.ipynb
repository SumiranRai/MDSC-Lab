## Data Understanding and Preparation

A. Read both the Datasets ‘Data1’ and ‘Data 2’ as DataFrame and store them into two
separate variables

```python
import pandas as pd
import seaborn as sns

df1 = pd.read_csv('Data1.csv')
df2 = pd.read_csv('Data2.csv')
```

B. Print shape and Column Names and DataTypes of both the Dataframes.

```python
df1.info()
df1.shape
```

```python
df2.info()
df2.shape
```

C. Merge both the Dataframes on ‘ID’ feature to form a single DataFrame

```python
merged_df = pd.merge(df1, df2, on='ID')
merged_df.info()
merged_df.shape
```

D. Change Datatype of below features to ‘Object’.


‘CreditCard’, ‘InternetBanking’, ‘FixedDepositAccount’, ‘Security’, ‘Level’, ‘HiddenScore’.


(Reason behind performing this operation:- Values in these features are binary i.e. 1/0. But DataType is ‘int’/’float’ which is not expected.)

```python
merged_df.head(15)
```

```python
merged_df['CreditCard'] = merged_df['CreditCard'].astype(str)
merged_df['InternetBanking'] = merged_df['InternetBanking'].astype(str)
merged_df['FixedDepositAccount'] = merged_df['FixedDepositAccount'].astype(str)
merged_df['Security'] = merged_df['Security'].astype(str)
merged_df['Level'] = merged_df['Level'].astype(str)
merged_df['HiddenScore'] = merged_df['HiddenScore'].astype(str)

merged_df.info()
```

## Data Exploration and Analysis

A. Visualize distribution of Target variable ‘LoanOnCard’ and clearly share insights.

```python
ax = sns.countplot(x='LoanOnCard', data=merged_df)

ax.set_title('Distribution of Loan on Card', fontsize=16)
ax.set_xlabel('Loan on Card (0 = No, 1 = Yes)', fontsize=12)

loan_counts = merged_df['LoanOnCard'].value_counts()
print(loan_counts)
```

### Inference

It is observed that out of 5000 Bank Customers, only 480 Customers have taken Loan on Card. 

Remaining 4500 haven't took any Loan on Card.

Data of 20 Bank Customers is Missing.

B. Check the percentage of missing values and impute if required.


```python
null_counts = merged_df.isnull().sum()

print("Count of null values in each column -")
print(null_counts)
```

```python
missing_percentage = (null_counts / 5000) * 100
print(missing_percentage)
```

```python
# Calculate the mode of the column
mode_value = merged_df['LoanOnCard'].mode()[0]

# Replace NaN values with the mode
merged_df['LoanOnCard'] = merged_df['LoanOnCard'].fillna(mode_value)

null_counts = merged_df.isnull().sum()

print("Count of null values in each column -")
print(null_counts)
```

C. Check for unexpected values in each categorical variable and impute with best suitable
value.

[Unexpected values means if all values in a feature are 0/1 then ‘?’, ‘a’, 1.5 are unexpected values which needs treatment]

```python
merged_df.head(15)
```

```python
# Converting float to int
merged_df['LoanOnCard'] = merged_df['LoanOnCard'].astype(int)

expected_values = {'0', '1'}

# Identify unexpected values
columns = ['Security', 'FixedDepositAccount', 'InternetBanking', 'CreditCard', 'LoanOnCard'] 
unique_values = set()

for col in columns:
    unique_values.update(merged_df[col].astype(str))

print(unique_values)
unexpected_values = unique_values - expected_values

print(f"Unexpected values: {unexpected_values}")

# Impute unexpected values with the mode (most frequent value)
for col in columns:
    mode_value = merged_df[col].mode()[0]
    merged_df[col] = merged_df[col].replace(unexpected_values, mode_value)
```
