Calculating the mean 'Age'.

```python
# Mean
def calculate_mean(df, column):
    total_sum, count = 0, 0
    for value in df[column]:
        total_sum += value
        count += 1
    mean = total_sum / count
    return mean

x = calculate_mean(df, 'AGE')
print('Mean Age is', x)
```

```python
# Median
def calculate_median(nums):
    sorted_nums = sorted(nums)
    n = len(sorted_nums)
    middle = n // 2
    if n % 2 == 0:
        return (sorted_nums[middle - 1] + sorted_nums[middle]) / 2
    else:
        return sorted_nums[middle]

# Mode
def calculate_mode(nums):
    frequency = {}
    for num in nums:
        frequency[num] = frequency.get(num, 0) + 1
    mode = max(frequency, key=frequency.get)
    return mode

# Range
def calculate_range(nums):
    return max(nums) - min(nums)

# Variance
def calculate_variance(nums):
    mean = calculate_mean(nums)
    squared_diff = [(x - mean) ** 2 for x in nums]
    return sum(squared_diff) / len(nums)

# Standard deviation
def calculate_std_dev(nums):
    variance = calculate_variance(nums)
    return variance ** 0.5
```

Transform data into a format suitable for plotting

```python
# Convert 'LUNG CANCER' values to 2 and 1
df1 = df
df1['LUNG_CANCER'] = df1['LUNG_CANCER'].replace({'YES': 2, 'NO': 1})

df1.info()

df1.head()
```

```python
import matplotlib.pyplot as plt

# For Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='GENDER', palette='dark', alpha=.6)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Convert 'GENDER' values to 2 and 1
df1['GENDER'] = df1['GENDER'].replace({'M': 1, 'F': 2})
df1 =df1.drop(columns=['GENDER'])
df1 = df1.drop(columns=['AGE'])
```

```python
# Transform data into a format suitable for plotting
df_melted = df1.melt(var_name='attribute', value_name='presence')

# Replace values with meaningful labels
df_melted['presence'] = df_melted['presence'].map({1: 'Absent', 2: 'Present'})

# Draw the barplot
g = sns.catplot(
    data=df_melted, kind="count",
    x="attribute", hue="presence",
    palette="dark", alpha=.6, height=8, aspect=2
)
g.despine(left=True)
g.set_axis_labels("Attribute", "Count")
g.legend.set_title("Presence")

plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability
plt.show()
```
