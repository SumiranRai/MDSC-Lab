# Plot -- Figure and Axes

Plotting a Simple Curve

Importing the libraries

```python
import matplotlib.pyplot as plt
import numpy as np 
x = np.linspace(0, 10, 100)#Create an array of 100 evenly sized values between 0 to 100
print(x)
```

Generating the data

```python
y = np.sin(x)# Calculate the sine value of each case
print(y)
```

```python
z = np.cos(x)#Calculate the cosine value of each case 
print(z)
```

Plotting the data

```python
plt.plot(x,y, label = "sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Simple Line Plot")
plt.legend()
plt.show()
```

Plotting Multiple Curves on a Plot

```python
plt.plot(x, y, label = "sin(x)")
plt.plot(x, z, label = "cos(x)")
plt.xlabel("x")
plt.ylabel("Value")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

# Contour Plots

Contour(Curved Lines) Plots are graphical represntations where contour lines are used to represent 3D data in 2D

Each contour line represents a constant value.

Applications -
Meteorology, Geography & Engineering


Generating Data for Contour Plots Using np.meshgrid()

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

Plotting the Contour Plots

```python
plt.contour(X, Y, Z, levels = 20)
plt.title("Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()
plt.show()

plt.contourf(X, Y, Z, levels = 20, cmap = "coolwarm")
#Creates filled contour plots
plt.title("Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()
plt.show()
```
